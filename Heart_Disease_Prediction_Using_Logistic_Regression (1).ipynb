```python
dataset['education'] = dataset.fillna('education').mean(axis = 1)
dataset['BPMeds'] = dataset.fillna('BPMeds').mean(axis = 1)
dataset['cigsPerDay'] = dataset.fillna('cigsPerDay').mean(axis = 1)
dataset['totChol'] = dataset.fillna('totChol').mean(axis = 1)
dataset['BMI'] = dataset.fillna('BMI').mean(axis = 1)
dataset['heartRate'] = dataset.fillna('heartRate').mean(axis = 1)
dataset['glucose'] = dataset.fillna('glucose').mean(axis = 1)

dataset.isnull().sum()
dataset.head()
```

```python
X = dataset.drop('TenYearCHD', axis = 1)
y = dataset['TenYearCHD']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test, y_train,y_test = train_test_split(X,y,test_size = 0.3, random_state = 42)
```

```python
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
X_train = scale.fit_transform(X_train)
X_test = scale.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
log_model = LogisticRegression()
log_model.fit(X_train,y_train)
```

```python
y_pred = log_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, classification_report
```

```python
print('Accuracy_score', accuracy_score(y_test, y_pred))
```

```python
log_model.score(X_train,y_train)
log_model.score(X_test,y_test)
```

```python
print('Confusion_matrix', confusion_matrix(y_test,y_pred))
```

```python

print("Classification Report \n\n",classification_report(y_test,y_pred))
```

```python
from sklearn.metrics import plot_confusion_matrix
plot_confusion_matrix(log_model,X_test, y_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn_model = KNeighborsClassifier(n_neighbors=5)
knn_model.fit(X_train,y_train)
```

```python
knn_pred = knn_model.predict(X_test)
```

```python
print('Test set score \n',knn_model.score(X_test,y_test))
print('Training set score\n',knn_model.score(X_train,y_train))
```

```python
plot_confusion_matrix(knn_model,X_test, y_test)
```
